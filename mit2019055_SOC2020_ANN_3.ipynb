```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
input_1 = np.array([0,0,1,1])
input_2 = np.array([0,1,0,1])
output_or = np.array([0,1,1,1])
count = 0
x = np.vstack((np.ones(4), input_1, input_2)).T
w = np.random.rand(1,3)
while(count != 4):
    count = 0
    for i in range(4):
      Y_out = np.dot(x[i], w.T)
      if (output_or[i] == 1 and Y_out < 0):
        w = w + x[i].T
      elif (output_or[i] == 0 and Y_out >= 0):
        w = w - x[i].T
      else :
        count += 1 
```

```
a = np.linspace(-0.5, 1.0, 100)
scatter = plt.scatter(input_1,input_2,c=output_or)
m = -w[0,1]/w[0,2]
c = -w[0,0]/w[0,2]
plt.plot(a,m*a + c, color = 'g')
plt.xlabel('input_1')
plt.ylabel('input_2')
plt.legend(*scatter.legend_elements(),loc="upper left", title="Classes")
plt.title('Or gate using perceptron')
plt.show()
```
